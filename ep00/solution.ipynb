```python
import jax
import jax.numpy as jnp
```

## Arrays and shapes

Same as NumPy: `jnp.array`, `jnp.arange`, `.shape`, `.dtype`.

```python
x = jnp.array([1.0, 2.0, 3.0, 4.0])
m = jnp.arange(12).reshape(3, 4)

print("x:", x.shape, x.dtype)
print("m:", m.shape)
print(m)
```

## Shapes

| Array | Shape |
|-------|-------|
| `x` | `(N,)` â€” vector of length `N` |
| `m` | `(I, J)` â€” matrix |

## Indexing

Identical to NumPy.

```python
print("row 0:", m[0])
print("cell [1,2]:", m[1, 2])
```

## Immutability

Unlike NumPy, you cannot write `x[0] = 99` â€” JAX raises `TypeError: JAX arrays are immutable`.

Updates return a **new** array via `.at[...].set(...)`. The original is untouched. (This matters once we `jit` and differentiate.)

```python
try:
    x[0] = 99.0
except TypeError as e:
    print(type(e).__name__, ":", e)

x_new = x.at[0].set(99.0)
print("x:    ", x)
print("x_new:", x_new)
```
